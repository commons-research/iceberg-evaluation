```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("results/evaluation_results.tsv", sep="\t")
```

```python
sns.violinplot(
    data=df,
    x="instrument_type",
    y="dreams_embedding_cosine_similarity",
    hue="adduct",
).set(
    xlabel="Instrument Type",
    ylabel="Dreams Embedding Cosine Similarity",
    title="Iceberg prediction performance by instrument type and adduct",
    ylim=(-0.05, 1.1),
)
```

```python
sns.violinplot(
    data=df,
    x="instrument_type",
    y="CosineGreedy_score",
    hue="adduct",
).set(
    xlabel="Instrument Type",
    ylabel="Cosine Greedy Score",
    title="Iceberg prediction performance by instrument type and adduct",
    ylim=(-0.05, 1.1),
)
```

```python
sns.violinplot(
    data=df,
    x="instrument_type",
    y="CosineHungarian_score",
    hue="adduct",
).set(
    xlabel="Instrument Type",
    ylabel="Cosine Hungarian Score",
    title="Iceberg prediction performance by instrument type and adduct",
    ylim=(-0.05, 1.1),
)
```

```python
sns.histplot(
    data=df.dropna(subset="collision_energy"),
    x="collision_energy",
    y="dreams_embedding_cosine_similarity",
    hue="adduct",
)
```

```python
sns.histplot(
    data=df, x="precursor_mz", y="dreams_embedding_cosine_similarity", cbar=True
)
```

```python
sns.scatterplot(
    data=df,
    x="CosineHungarian_score",
    y="dreams_embedding_cosine_similarity",
    alpha=0.8,
    s=3,
    hue="adduct",
)
```
